````python
import numpy as np
import pandas as pd

df = pd.read_csv('tested.csv')
print(df.shape)
df.head()
````

````python
def my_mean(values):
    return sum(values) / len(values)

def my_variance(values):
    mean = my_mean(values)
    return sum((x - mean) ** 2 for x in values) / len(values)

def my_std(values):
    return my_variance(values) ** 0.5
````

ages = df['Age'].dropna().tolist()

print("=== MY SCRATCH FUNCTIONS ===")
print(f"Mean Age:     {my_mean(ages):.4f}")
print(f"Variance Age: {my_variance(ages):.4f}")
print(f"Std Age:      {my_std(ages):.4f}")

print("\n=== NUMPY VERIFICATION ===")
print(f"Mean Age:     {np.mean(ages):.4f}")
print(f"Variance Age: {np.var(ages):.4f}")
print(f"Std Age:      {np.std(ages):.4f}")

print("\n✅ Match!" if round(my_std(ages), 4) == round(np.std(ages), 4) else "❌ Check your code")
